```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing,  svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/194903-202105_Bottle.csv', encoding='latin-1')
df2 = df[['ChlorA', 'PO4uM']] # Mencari hubungan antara kadar klorofil dengan fosfat

df2.columns = ['Chlorophyl', 'Phosphate']
new_df2 = df2.dropna() # Membersihkan data dengan menghilangkan nilai null
print(df2.shape) # Memerika jumlah baris sebelum dibersihkan (895371 baris)
print(new_df2.shape) # Memeriksa jumlah baris yang tersisa setelah dibersihkan (234182 baris)
```

```python
new_df2.head()
```

```python
sns.lmplot(x ="Chlorophyl", y ="Phosphate", data = new_df2,  order = 2, ci  = None)
plt.show()
```

```python
new_df2.corr()
# Angka -0.184424 menunjukkan bahwa kedua variabel memiliki korelasi yang berbanding terbalik namun hampir tak terlihat adanya
```

```python
X = np.array(new_df2['Chlorophyl']).reshape(-1, 1)
y = np.array(new_df2['Phosphate']).reshape(-1, 1)

new_df2.dropna(inplace = True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

regr = LinearRegression()

regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

```python
y_pred = regr.predict(X_test)
plt.scatter(X_test, y_test, color ='b' )
plt.plot(X_test, y_pred,  color ='k')

plt.show()
```

```python
X = np.array(new_df2['Chlorophyl']).reshape(-1, 1)
y = np.array(new_df2['Phosphate']).reshape(-1, 1)

new_df2.dropna(inplace = True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

regr = LinearRegression()

regr.fit(X_train, y_train)

y_pred = regr.predict(X_test)
plt.scatter(X_test, y_test, color ='b' )
plt.plot(X_test, y_pred,  color ='k')

plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
mse = mean_squared_error(y_true=y_test,y_pred=y_pred)
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred,squared=False)

print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```
